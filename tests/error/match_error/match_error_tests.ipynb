```python
# Minimal valid match statement (should compile)
from guppylang import guppy

@guppy.enum
class Direction:
    North = {"A": int}
    South = {"B": int}
    East = {}
    West = {}

@guppy
def main(north: Direction) -> int:
    match north:
        case Direction.North(1):
            return 1
        case Direction.South:
            return 2
        case _:
            return 0

main.check()
```
